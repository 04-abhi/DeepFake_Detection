```python
import os
import shutil

# Define the paths to the dataset and the destination directories
dataset_path = 'dataset'  # Update this to your dataset path
output_path = 'dataset'  # Update this to your desired output path

# Paths to the original and manipulated sequences
original_sequences_path = os.path.join(dataset_path, 'original_sequences', 'youtube', 'c23', 'videos')
manipulated_sequences_path = os.path.join(dataset_path, 'manipulated_sequences')

# Output directories for real and fake videos
real_videos_path = os.path.join(output_path, 'real')
fake_videos_path = os.path.join(output_path, 'fake')

# Create the output directories if they don't exist
os.makedirs(real_videos_path, exist_ok=True)
os.makedirs(fake_videos_path, exist_ok=True)

# Copy original (real) videos
for video_file in os.listdir(original_sequences_path):
    src_file = os.path.join(original_sequences_path, video_file)
    dst_file = os.path.join(real_videos_path, video_file)
    shutil.copy2(src_file, dst_file)
    print(f'Copied real video: {video_file}')

# Copy manipulated (fake) videos from each manipulation method
manipulation_methods = ['Deepfakes', 'Face2Face', 'FaceSwap', 'NeuralTextures']
for method in manipulation_methods:
    method_path = os.path.join(manipulated_sequences_path, method, 'c23', 'videos')
    for video_file in os.listdir(method_path):
        src_file = os.path.join(method_path, video_file)
        dst_file = os.path.join(fake_videos_path, f'{method}_{video_file}')
        shutil.copy2(src_file, dst_file)
        print(f'Copied fake video: {method}_{video_file}')

print('Dataset organization complete.')
```
